```python
import scanpy as sc
import pandas as pd
import gc
import matplotlib.pyplot as plt
import pickle
from samalg import SAM
from samap.mapping import SAMAP
from samap.analysis import get_mapping_scores, sankey_plot
```

# Load in and preprocess the data

```python
mlig = sc.read_h5ad( 'ArchROutputs/Mlig/Mlig.genescores_init.h5ad' )
mlig.obs['Sample'] = [ bc.split('#')[0] for bc in mlig.obs_names ]
# only take uncut cells to save memory
mlig = mlig[mlig.obs.Sample=='uncut',:].copy()

mlig
```

```python
samM = SAM( counts=mlig )
samM.preprocess_data()
samM.run( weight_mode='rms', sparse_pca=True, batch_key='Sample' )
```

```python
del mlig
gc.collect()
```

```python
fig = plt.figure( figsize=(12,8) )

sc.pl.umap( samM.adata, color='SeuratClusters', size=50, ax=fig.gca(), legend_loc='on data' )
```

```python
samM.clustering( param=2.1 )
```

```python
smed = sc.read_h5ad( 'ArchROutputs/Smed/Smed.genescores_init_imputed.h5ad' )
smed.obs['Sample'] = [ bc.split('#')[0] for bc in smed.obs_names ]

smed
```

```python
samP = SAM( counts=smed )
# Imputed gene scores for Smed are still pretty sparse
samP.preprocess_data( filter_genes=False )
samP.run( weight_mode='rms', sparse_pca=True, batch_key='Sample' )
```

```python
del smed
gc.collect()
```

```python
fig = plt.figure( figsize=(12,8) )

sc.pl.umap( samP.adata, color='SeuratClusters', size=50, ax=fig.gca(), legend_loc='on data' )
```

```python
samP.clustering( param=0.8 )
```

```python
sman = sc.read_h5ad( 'ArchROutputs/Sman/Sman.genescores_init.h5ad' )
sman.obs['Sample'] = [ bc.split('#')[0] for bc in sman.obs_names ]
# reformat the names since ArchR didn't like the underscores
sman.var_names = [ gene.replace('-','_') for gene in sman.var_names ]

sman
```

```python
samS = SAM( counts=sman )
samS.preprocess_data()
samS.run( weight_mode='rms', sparse_pca=True, batch_key='Sample' )
```

```python
del sman
gc.collect()
```

```python
fig = plt.figure( figsize=(12,8) )

sc.pl.umap( samS.adata, color='SeuratClusters', size=50, ax=fig.gca(), legend_loc='on data' )
```

```python
samS.clustering( param=1.2 )
```

# Run SAMap

```python
samap = SAMAP( { 'Ml': samM, 'Pl': samP, 'Sm': samS }, f_maps='samap_genemaps/' )
```

```python
mapping = samap.run(pairwise=True)
```

```python
fig = plt.figure( figsize=(12,8) )

sc.pl.umap( mapping.adata, color='species', size=50, ax=fig.gca() )
```

# Compare clusters

```python
keys = { 'Ml':'leiden_clusters', 'Pl':'leiden_clusters', 'Sm':'leiden_clusters' }
D, MappingTable = get_mapping_scores( samap, keys, n_top = 0 )
```

```python
sankey_plot( MappingTable, align_thr=0.01, species_order = ['Pl','Ml','Sm'] )
```

```python
sankey_plot( MappingTable, align_thr=0.01, species_order = ['Ml','Pl','Sm'] )
```

```python
sankey_plot( MappingTable, align_thr=0.01, species_order = ['Ml','Sm','Pl'] )
```

At this point, we manually merged some clusters based on the mappings, and then read back in the new annotations

```python
samap.sams['Ml'].adata.obs['GroupFigure'] = pd.read_csv( 'Metadata/Mlig.final_cluster_annots.csv', index_col=0 )\
                                                .loc[samap.sams['Ml'].adata.obs_names,'GroupFigure']
samap.sams['Pl'].adata.obs['GroupFigure'] = pd.read_csv( 'Metadata/Smed.final_cluster_annots.csv', index_col=0 )\
                                                .loc[samap.sams['Pl'].adata.obs_names,'GroupFigure']
samap.sams['Sm'].adata.obs['GroupFigure'] = pd.read_csv( 'Metadata/Sman.final_cluster_annots.csv', index_col=0 )\
                                                .loc[samap.sams['Sm'].adata.obs_names,'GroupFigure']
```

```python
keys = { 'Ml':'leiden_clusters', 'Pl':'leiden_clusters', 'Sm':'leiden_clusters' }
D, MappingTable = get_mapping_scores( samap, keys, n_top = 0 )
```

```python
MappingTable.to_csv( 'Metadata/GroupFigure_SAMap_scores.csv' )
```
