```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
import seaborn as sns
```

```python
## Mouse and zebrafish
zebrafish = pd.read_csv("zebrafish_metrics.csv",sep=',')
mouse = pd.read_csv("mouse_metrics.csv", sep=',')

FOLDS  = 3

zebrafish_fold = zebrafish.fold.unique()
mouse_fold = mouse.fold.unique()
```

```python
plt.rcParams['figure.figsize'] = 3,3

family = ['neural', 'cardio', 'endo']
region = ['forebrain', 'hindbrain', 'spinal']
celltype = ['optics','mid', 'posterior']

zebra_family = zebrafish[zebrafish['cell type'].isin(family)]
zebra_region = zebrafish[zebrafish['cell type'].isin(region)]
zebra_type = zebrafish[zebrafish['cell type'].isin(celltype)]


fig, axes = plt.subplots(1,3, figsize=(12,4), sharey=True)

sns.barplot(
    x='cell type',
    y='pearson_r',
    data=zebra_family,
    hue='fold',
    ax=axes[0]
)

axes[0].set_title('zebrafish (cell type families)')
axes[0].set_ylim([0,0.85])

sns.barplot(
    x='cell type',
    y='pearson_r',
    data=zebra_region,
    hue='fold',
    ax=axes[1]
)

axes[1].set_title('zebrafish (brain regions)')
axes[1].set_ylim([0,0.85])

plt.tight_layout()
plt.ylim([0,0.85])

sns.barplot(
    x='cell type',
    y='pearson_r',
    data=zebra_type,
    hue='fold',
    order=celltype,
    ax=axes[2]
)

axes[2].set_title('zebrafish (cell type)')
axes[2].set_ylim([0,0.85])

plt.tight_layout()
plt.ylim([0,0.85])

#plt.show()


#sns.barplot(x='cell type',y='pearson_r',data=zebrafish,hue='fold')
plt.savefig("./Figures/zebrafish_pearson_r_fold_histogram_updated.svg",format='svg')
```
