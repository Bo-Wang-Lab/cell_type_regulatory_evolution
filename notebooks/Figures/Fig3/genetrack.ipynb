### Loading cell type annotation

```r
cellnames <- read.csv("./Metadata/Sman.obs_table_singlecell.csv", sep = ",")
rownames(cellnames) <- cellnames$X
Sman<- addCellColData(Sman,data = cellnames$GroupFigure, name = 'GroupFigure', cells=rownames(cellnames), force=TRUE)
Sman<- addCellColData(Sman,data = cellnames$POU_label, name = 'POU_label', cells=rownames(cellnames), force=TRUE)
```

```r
cellnames <- read.csv("./Metadata/Mlig.obs_table_singlecell.csv", sep = ",")
rownames(cellnames) <- cellnames$X
Mlig<- addCellColData(Mlig,data = cellnames$GroupFigure, name = 'GroupFigure', cells=rownames(cellnames), force=TRUE)
Mlig<- addCellColData(Mlig,data = cellnames$POU_label, name = 'POU_label', cells=rownames(cellnames), force=TRUE)
```

```r
cellnames <- read.csv("./Metadata/Smed.obs_table_singlecell.csv", sep = ",")
rownames(cellnames) <- cellnames$X
Smed<- addCellColData(Smed,data = cellnames$GroupFigure, name = 'GroupFigure', cells=rownames(cellnames), force=TRUE)
Smed<- addCellColData(Smed,data = cellnames$POU_label, name = 'POU_label', cells=rownames(cellnames), force=TRUE)
```
