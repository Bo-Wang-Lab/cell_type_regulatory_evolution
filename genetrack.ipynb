## Hv-pou4

```r
## This is for hydra 

genomeInfo <- read.table("/Refs/HVAEP/HVAEP.chr.chrom.sizes",stringsAsFactors = F)

chrominfo <- data.frame(chrom=genomeInfo$V1, length=genomeInfo$V2, is_circular=rep(FALSE,length(genomeInfo$V1)))

txDb.mi <- makeTxDbFromGFF(file="/Refs/HVAEP/HVAEP.GeneModels_fixed.gff3", format="auto", dataSource="Gene Models", organism ="Hydra vulgaris", chrominfo=chrominfo)

hydra_peaks <- getPeaks("/FilteredPeakMotifHits/Hv.filt_peak_motif_hits.narrowPeak",sort_peaks = FALSE)

options(ucscChromosomeNames=F)

AtacPlot <- function(chr, height, left, right, buffer, fpath = "plot.pdf") {
  chrID <- chr
  
  #This specifies the maximum value on the y-axis
  z <- height
  
  # Chromosome Map
  gtrack <- GenomeAxisTrack(name = chrID,add53=T, add35 = T, fontsize = 13, fontcolor.title = "black", fontsize.title = 13, showTitle = F, rotation.title = 0, grid = T,cex = 0.6, labelPos = "below")
  
  # Gene Models
  grtrack <- GeneRegionTrack(txDb.mi, chromosome=chrID, name="Genes", transcriptAnnotation = "gene", col = "black", fontsize.group = 13, fontcolor.group = "black", fill = "black", fontsize=25, rotation.title = 0, background.title = "white", col.line = "black",just.group = "below", collapseTranscripts = "longest")
  
  atrack <- AnnotationTrack(hydra_peaks, name = "peaks")

  # Data Tracks
  nGreen_all <- DataTrack(range = "/Coveragetracks/nGreen_PN_final_shift.bw", type = "polygon", col = "black", lwd.mountain = 0, fill.mountain = c("#E11584","#E11584"), chromosome = chrID, name = "Control", fontsize = 13, showAxis = T, ylim=c(0,z), background.title = "white", fontcolor.title = "black", col.axis = "black", span = 0.5, lwd = 0, showTitle = F, cex.axis = 0.6)
  
  AEP_all <- DataTrack(range = "/Coveragetracks/AEP_MG_final_shift.bw", type = "polygon", col = "black", lwd.mountain = 0, fill.mountain = c("#323232","#323232"), chromosome = chrID, name = "Control", fontsize = 13, showAxis = T, ylim=c(0,z), background.title = "white", fontcolor.title = "black", col.axis = "black", span = 0.5, lwd = 0, showTitle = F, cex.axis = 0.6)
  

  pdf(file = fpath, width=15, height=10)
  plotTracks(list(AEP_all, nGreen_all, atrack, grtrack, gtrack),from=(left - buffer), to=(right + buffer), title.width = 0.7, margin = 0, sizes = NULL)
  dev.off()
}
```
