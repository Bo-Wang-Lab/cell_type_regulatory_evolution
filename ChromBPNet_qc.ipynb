```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
import seaborn as sns
```

```python
## The model metrics were taken from the chrombpnet_metrics.json from each model and accumulated 
Mlig = pd.read_csv("./ChromBPNet/QC/Mlig_metrics.csv",sep=',')
Smed = pd.read_csv("./ChromBPNet/QC/Smed_metrics.csv", sep=',')
Sman = pd.read_csv("./ChromBPNet/QC/Sman_metrics.csv",sep=',')
Hv = pd.read_csv("./ChromBPNet/QC/Hv_metrics_new.csv",sep=',')
Nv = pd.read_csv("./ChromBPNet/QC/Nv_metrics.csv", sep=',')

Mlig['celltype'] = Mlig['celltype'].replace('GSC','gsc')
Smed['celltype'] = Smed['celltype'].replace('five_type','z')
Sman['celltype'] = Sman['celltype'].replace('GSC','gsc')
Sman['celltype'] = Sman['celltype'].replace('S1_vitellocyte', 's1')

FOLDS  = 3
Mlig = Mlig.sort_values(by='celltype',ascending=True)
Smed = Smed.sort_values(by='celltype',ascending=True)
Sman = Sman.sort_values(by='celltype',ascending=True)

Mlig_fold = Mlig.fold.unique()
Smed_fold = Smed.fold.unique()
Sman_fold = Sman.fold.unique()
Hv_fold = Hv.fold.unique()
Nv_fold = Nv.fold.unique()
```

## Mlig model QC

```python
sns.barplot(x='celltype',y='pearson_r',data=Mlig,hue='fold')
plt.ylim([0,0.85])
plt.savefig("./Plots/EDFig5/PanelED5c_Mligtop.svg",format='svg')
```

```python
sns.barplot(x='celltype',y='jsd',data=Mlig,hue='fold')
plt.ylim([0,0.85])
plt.savefig("./Plots/EDFig5/PanelED5c_Mligbottom.svg",format='svg')
```
